```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
df
```

```python
df.head(2)
```

```python
df["species"]= df["species"].map({"setosa":0,"versicolor":1,"virginica":2})
```

```python
df
```

```python
df["species"]= df["species"].replace({0:"zero",1:"one",2:"two"})
```

```python
df
```

```python
#df["species"].astype("int")
```

```python
pd.get_dummies(df["species"],drop_first=True)
```

```python
df1=pd.get_dummies(df["species"],drop_first=True)
```

```python
df=pd.concat([df,df1],axis=1)
df
```

```python
df
```

```python
df["sq_sepal_length"]=df["sepal_length"].apply(lambda x:x**2)
df
```

```python
df["len_sp"]=df["species"].apply(lambda x:len(x))
```

```python
df["even_odd"]=df["petal_width"].apply(lambda x:x//2==0)
```

```python
df["sq_sepal_length"]=df["sq_sepal_length"].astype("int")
```

```python
df
```

```python
df["new"]=df["sq_sepal_length"].apply(lambda x:x%2==0)
```
