```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,5))
plt.axis()
```

```python
%matplotlib inline

x1=np.linspace(0,10,100)
#print(x1)
#create a plot figure
fig=plt.figure()
plt.plot(x1,np.sin(x1),"--k")  #--k doted line with black colour
plt.plot(x1,np.cos(x1),"or");    #red colour
```

```python
#to see different plots at a time  : subplot
```

```python
plt.figure(figsize=(10,10))

#create the first of two panels and set current axis
plt.subplot(1,2,2)
plt.plot(x1,np.sin(x1))

#create the second of two panels and set current axis
plt.subplot(1,2,1)
plt.plot(x1,np.cos(x1))
```

```python
plt.plot([1,8,10,5,6],[1,2,3,4,5])
plt.ylabel("Numbers")
plt.xlabel("x variable")
plt.title("List plot on matplotlib")
plt.xticks(np.arange(1,11,1))
plt.yticks(np.arange(1,10,2))
plt.show();
```

```python
x=np.linspace(0,2,100)

plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")

plt.xlabel("x label")
plt.ylabel("y label")

plt.title("Simple Plot")

plt.legend()
plt.show()
```

```python
t=np.linspace(0,2,100)
plt.plot(t,t,"r--",t,t**2,"bs",t,t**3,"g^")
plt.show();
```

```python
x3=np.arange(0.0,6.0,0.01)
plt.plot(x3,[i**2 for i in x3])
plt.show()
```

# Error Bar Chart

```python
from numpy import random
```

```python
x9=np.arange(0,4,0.2)
print(x9)
print(len(x9))
y9=np.exp(-x9)
y9
print(len(y9))
s=np.random.randn(len(y9))
np.abs(s)
```

```python
x9=np.arange(0,4,0.2)
y9=np.exp(-x9)
e1=0.1*np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt="o-")
plt.show();
```
